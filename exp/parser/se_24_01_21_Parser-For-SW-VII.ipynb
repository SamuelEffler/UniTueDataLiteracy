# Parser - SW Episode VII

The use of this parser is to extract *spoken words* and the talking *character's name* from given movie scripts as HTML-files.
The goal is a SVG-file containing columns for *character's name*, *spoken phrase*, and optional *number of words*.


## Implementation Plan

The algorithm is taken from the other Parser (v02 for Episode IV) and modified to fit Episode VII

```python
# Imports and Setup

import csv
import re

# Parameters for 'Star Wars: A New Hope'

# Number of whitespaces between '<b>' and CHARACTER NAME
wsChar = 26
wsAdd = 11
wsPhrase = 11
```

```python
# Create a CSV
header = ['name', 'phrase', 'numberOfWords']

data = open('./output/swvii.csv', 'w', encoding='UTF-8', newline='')

writer = csv.writer(data)

writer.writerow(header)

# Create a TXT file with all spoken words
words = open('./output/spokenWordsSWVII.txt', 'a', encoding='UTf-8')

```

### Regular Expressions

```python
# Name line (Name in Group)
# '^<b>\s{37}(?P<name>[^\s].{1,})'gm

# Phrase line (Phrase in Group)
# '^(</b>)?\s{25}(?P<phrase>[^\s].{1,})'gm

# End of phrase line
# '^$'gm
```

## Parser v01

```python
# Parser for Episode VII

# Names to find
names = ['HAN', 'LEIA', 'KYLO REN', 'REY', 'FINN', 'POE', 'MAZ', 'SNOKE', 'GENERAL HUX', 'C-3PO']

# Open file
with open('./filmScripts/Star-Wars-The-Force-Awakens.html', 'r') as script:
    
    for line in script:

        m = re.findall(r'^<b>\s{11,}(?P<name>' + '|'.join(names) + ')', line)
        if m:

            # Save name
            name = m[0]
            
            # Save phrase
            phrase = ''
            while not re.match(r'^$', line):
                line = next(script)
                if re.match(r'.*\(', line):
                    while not re.match(r'.*\)', line):
                        line = next(script)
                    line = next(script)
                m = re.match(r'^(</b>)?\s{11}(?P<phrase>[^\s].{1,})', line)
                if m:
                    phrase = phrase + m.group('phrase') + ' '
                    phrase = re.sub(r'[,\.!?:]|"|--(\s)?', '', phrase)
            phrase = re.sub(r' $', '', phrase)

            # Count words
            numberOfWords = len(phrase.split())

            # Write svg
            if numberOfWords > 0:
                writer.writerow([name, phrase, numberOfWords])

            # Write txt
                phrase = phrase + ' \n'
                words.writelines(phrase)
```

## Close Files - Run always!

```python
# Close all open files
data.close()
words.close()
```
