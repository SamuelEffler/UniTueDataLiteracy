## Episode IV - Parser v02

```python
# Open file
with open('./filmScripts/Star-Wars-A-New-Hope.html', 'r') as script:
    
    for line in script:

        m = re.match(r'^<b>\s{37}(?P<name>[^\s].{1,})', line)
        if m:

            # Save name
            name = m.group('name')

            # Replace Vocal Names for the droid's numbers
            name = re.sub(r'THREEPIO', 'C3PO', name)
            
            # Save phrase
            phrase = ''
            while not re.match(r'^$', line):
                line = next(script)
                m = re.match(r'^(</b>)?\s{25}(?P<phrase>[^\s].{1,})', line)
                if m:
                    phrase = phrase + m.group('phrase')
                    phrase = re.sub(r'[,\.!?]|-- ', '', phrase)

                    # Replace Vocal Names for the droid's numbers
                    phrase = re.sub(r'Artoo-Detoo|Artoo', 'R2D2', phrase)
                    phrase = re.sub(r'Threepio', 'C3PO', phrase)

            # Count words
            numberOfWords = len(phrase.split())

            # Write svg
            writer.writerow([name, phrase, numberOfWords])

            # Write txt
            phrase = phrase + ' \n'
            words.writelines(phrase)
```
